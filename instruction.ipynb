This project implements a RAG pipeline for arXiv Computational Linguistics papers, including data collection, text processing, embedding generation, FAISS indexing, and a
FastAPI service.

Task 1 & 2: Data Collection & Text Extraction

python pdf_extraction.py

This will:
Download 50 arXiv cs.CL papers via the arXiv API
Extract text from PDFs using PyMuPDF
Clean and save text to arxiv.json

Task 3: Text Chunking

python chunking.py

This will:
Load papers from papers_text.json
Split text into 512-token chunks with 50-token overlap

Task 4 & 5: Embedding Generation & FAISS Indexing

python embedding_faiss.py

This will:
Generate embeddings using sentence-transformers
Build FAISS index (IndexFlatL2)
Save index and chunks

Task 6: Test with Jupyter Notebook

jupyter notebook demo.ipynb

The notebook includes:
Interactive search interface
Example queries
Context retrieval
Index statistics

Task 7: Run FastAPI Service

Access the API at:
API: http://localhost:8000
Docs: http://localhost:8000/docs
Health check: http://localhost:8000/health
