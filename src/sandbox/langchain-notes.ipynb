```python
from langchain.agents import create_agent
from langchain_ollama import ChatOllama

# import the tools
from tools.text_cleaning import text_cleanser

# define the model
llm = ChatOllama(
    model="smollm2:1.7b", # supports tool calling
    temperature=0
)

agent = create_agent(
    model = llm,
    tools=[text_cleanser],
    system_prompt="You are a helpful assistant."
)

agent.invoke(
    {
        "messages": [
            {
                "role":"user",
                "content": "Use the text cleanser tool to clean the words: 'HELLO World! " 
                + "Check out my site: https://example.com. It's awesome!'. "
                + "Not case sensitive. Just return the cleaned text"
            }
        ]
    }
)
```

```python
from langchain_google_genai import GoogleGenerativeAI
from getpass import getpass
from dotenv import load_dotenv
import os

load_dotenv()

model_name = 'gemini-2.5-flash'

# Get from env
if os.getenv('GEMINI_API_KEY'):
    api_key = os.getenv('GEMINI_API_KEY')
else:
    api_key = getpass()

google_llm = GoogleGenerativeAI(model=model_name, google_api_key=api_key)

result = google_llm.invoke(
    "Whats the best programming language in the market other than Python?"
)
```

```python
print(result)
```

```python
# lets use chains
from langchain_core.prompts import PromptTemplate

template = """Question: {question}

Answer: Lets think about this step by step."""

# Load the template
prompt = PromptTemplate.from_template(template)

# create a chain
chain = prompt | llm

question = "How much is 3 + 4 - 9? Explain your thought process."
smolm2_answer = chain.invoke({
    "question": question
})

print(smolm2_answer)
```

```python
chain2 = prompt | google_llm
question = "How much is 3 + 4 - 9 ? Explain your thought process."
google_answer = chain2.invoke({
    "question": question
})

print(google_answer)
```
