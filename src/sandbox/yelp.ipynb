```python
YELP_BASE_URL='https://api.yelp.com/v3'
YELP_BUSINESS_SEARCH_ENDPOINT = 'https://api.yelp.com/v3/businesses/search'
```

```python
import requests
from dotenv import load_dotenv
import getpass
import os

load_dotenv

api_key = os.environ.get("YELP_API_KEY")
location = '88 e san carlos street'
term = 'kenyan food'

# payload
payload = {
    'location': location,
    'term': term
}

# headers
headers = {
    'user-agent': 'my-hackathon-app/0.0.1',
    'authorization': f'Bearer {api_key}'
}

r = requests.get(YELP_BUSINESS_SEARCH_ENDPOINT, headers=headers, params=payload)
r.json()
```

```python
import json

if not r.status_code == 200:
    print("There seems to be a problem with your request.")
else:
    data = r.json()
    with open('data.json', 'w', encoding='utf-8') as file:
        json.dump(data, file, indent=4)
```

```python
for item in data['businesses']:
    print(f"{item['name']} - {item['rating']} - {item['display_address']} - {item['is_closed']}")
```

```python
for k, v in data.items():
    print(f"{k}\t:\t{v} ")
```
