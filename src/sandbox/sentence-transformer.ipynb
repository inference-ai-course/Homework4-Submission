```python
from sentence_transformers import SentenceTransformer

# The sentence(s) i want to transform to an embedding.
sentences = [
    "My name is Ian and I am Kenyan.",
    "I am a student in the USA",
    "I am currently working with sentence embeddings",
    "Kenya is an amazing place.",
    "I would love to explore the USA"
]

sentences_2 = [
    "Kenya is a beautiful country."
]

# we will use the "sentence-transformers/all-MiniLM-L6-v2" mdel
all_mini_model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

text_embeddings = all_mini_model.encode(sentences)
text_embeddings_2 = all_mini_model.encode(sentences_2)
```

```python
# lets find the similarity

how_similar = all_mini_model.similarity(text_embeddings, text_embeddings_2)
print(how_similar.shape)
print(how_similar)
```
