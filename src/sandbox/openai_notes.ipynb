```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI(
    model='gpt-5-mini',
    temperature=0
)

messages = [
    ("system", "You are a helpful assistant, Tim. You work at a tech store. You goal is to help a customer understand basic hardware question."),
    ("user", "How do you measure the size of a screen? What if it is not rectangular?")
]

ai_msg = llm.invoke(messages)
# print(ai_msg)

content = ai_msg.content
print(content)

```
