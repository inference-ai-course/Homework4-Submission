```python
def get_index(char)->int:
    FIRST_CHARACTER = 'A'
    # 1. Check if the input is a string
    if not isinstance(char, str):
        raise TypeError(f"Input must be a string. Received '{char}'")
    
    # 2. Check if the string length is 1
    if len(char) != 1:
        raise ValueError(f"Input must be a single character, but received a string of length {len(char)}.")
    
    if not char.isalpha():
        raise ValueError(f"Input character must be an English letter (A-Z or a-z). Received: '{char}'.")

    return ord(char) - ord(FIRST_CHARACTER)
```
